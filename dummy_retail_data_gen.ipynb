```python
# Helpers

def format_date(dt):
    return dt.strftime("%Y-%m-%d %H:%M:%S")

def random_date(start, end):
    delta = end - start
    return start + timedelta(seconds=random.randint(0, int(delta.total_seconds())))

def random_phone():
    return f"+{random.randint(1,999)}-{random.randint(100,999)} {random.randint(100,999)} {random.randint(1000,9999)}"

def random_email(first_name, last_name, used):
    first_name = first_name.lower()
    last_name = last_name.lower()
    email = f"{first_name}.{last_name}@{random.choice(DOMAINS)}"
    while email in used:
        email = f"{first_name}.{last_name}{random.randint(1,999)}@{random.choice(DOMAINS)}"
    used.add(email)
    return email

def get_last_order_id(volume_path):
    try:
        with open(volume_path+"/last_order_id.txt", "r") as f:
            return int(f.readline().strip())
    except:
        return 0   # No file, start from 0 id
    
def update_latest_order_id(orders, volume_path):
    max_order_id = max(orders, key=lambda x: x["order_id"])["order_id"]
    with open(volume_path+"/last_order_id.txt", "w") as f:
        f.write(str(max_order_id))
```

```python
# Generator

def gen_customers(n):
    customers = []
    used_email = set()
    now = datetime.now()
    two_years_ago = now - timedelta(days=365*2)

    for i in range(n):
        first_name = random.choice(FIRST_NAMES)
        last_name = random.choice(LAST_NAMES)
        email = random_email(first_name,last_name,used_email)
        city, country = random.choice(CITIES)
        phone = random_phone()
        address = f"{random.randint(1, 999)} {random.choice(STREETS)}, {city}, {country}"
        created_at = random_date(two_years_ago, now)
        
        customers.append({
            "customer_id": random.randint(1, 1000),
            "first_name": first_name,
            "last_name": last_name,
            "email": email,
            "phone": phone,
            "address": address,
            "created_at": format_date(created_at)
        })
    return customers


def gen_products(n):
    products = []
    now = datetime.now()
    two_years_ago = now - timedelta(days=365*2)

    for i in range(n):
        name, category, price, cost = PRODUCTS[i%len(PRODUCTS)]
        suffix = f" v{(i) // len(PRODUCTS)}" if i>=len(PRODUCTS) else ""
        created_at = random_date(two_years_ago, now)
        
        products.append({
            "product_id": i+1,
            "name": name+suffix,
            "category": category,
            "price": price,
            "cost": cost,
            "stock_quantity": random.randint(0, 500),
            "created_at": format_date(created_at)
        })

    return products


def gen_orders(n, products, customers,volume_path):
    orders = []
    now = datetime.now()
    two_years_ago = now - timedelta(days=365*2)
    last_order_id = get_last_order_id(volume_path)

    for i in range(n):
        customer = random.choice(customers)
        product = random.choice(products)
        customer_id = customer["customer_id"]
        product_id = product["product_id"]
        quantity = random.randint(1, 10)
        price = product["price"]
        status = random.choices(ORDER_STATUSES, weights=ORDER_STATUS_WEIGHTS)[0]
        payment = random.choices(PAYMENT_METHODS, weights=PAYMENT_WEIGHTS)[0]
        order_date = random_date(two_years_ago, now)

        orders.append({
            "order_id": i+1+last_order_id,
            "customer_id": customer_id,
            "product_id": product_id,
            "quantity": quantity,
            "price": price,
            "status": status,
            "payment": payment,
            "order_date": format_date(order_date)
        })

    return orders
```

```python
# Writers
def write_csv(records, volume_path, file_name, orders=False):
    if not records:
        return
    
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    filepath = volume_path+'/'+file_name+'_'+timestamp+'.csv' 

    with open(filepath, 'w', newline="") as f:
        writer = csv.DictWriter(f, fieldnames=records[0].keys())
        writer.writeheader()
        writer.writerows(records)
    print(f"  ✓ {filepath}  ({len(records):,} rows)")

    if orders:
        update_latest_order_id(records, volume_path)
```
